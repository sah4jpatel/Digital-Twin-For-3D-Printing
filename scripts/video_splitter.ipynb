```python
import cv2
import os
import numpy as np
```

```python
def extract_frames(video_path, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error: Could not open video.")
        return

    x_coordinates = [0, 0, 510, 1150, 1280, 1280]
    y_coordinates = [720, 375, 320, 350, 435, 720]
    frame_count = 0
    while True:
        success, frame = cap.read()

        if not success:
            break

        frame_name = f"Layer_{frame_count}.jpg"
        mask = np.zeros_like(frame)
        polygon = np.array(list(zip(x_coordinates, y_coordinates)), np.int32)
        cv2.fillPoly(mask, [polygon], (255, 255, 255))
        masked_image = cv2.bitwise_and(frame, mask)
        cv2.imwrite(os.path.join(output_folder, frame_name), masked_image)
        print(f"Saved {frame_name}")

        frame_count += 1

    cap.release()
    print("Done extracting frames.")

model = "AbstractArt"

video_path = f'../Dataset/{model}/timelapse.avi'
output_folder = f'../Dataset/{model}/Captures'
extract_frames(video_path, output_folder)
```

```python
model = "AbstractArt"
directory = f"C:/Users/sah4j/Downloads/Blender_Proj/Digital-Twin-For-3D-Printing/Dataset/{model}/Renders"

x_coordinates = [0, 0, 500, 1140, 1280, 1280]
y_coordinates = [720, 440, 415, 450, 590, 720]

for filename in os.listdir(directory):
    image = cv2.imread(os.path.join(directory, filename))
    width, height = image.shape[1], image.shape[0]
    new_width = int(width * 0.75)
    new_height = int(height * 0.75)
    left = (width - new_width) // 2
    top = (height - new_height) // 2
    right = left + new_width
    bottom = top + new_height
    cropped_image = image[top:bottom, left:right]
    resized_image = cv2.resize(cropped_image, (1280, 720))

    mask = np.zeros_like(resized_image)
    polygon = np.array(list(zip(x_coordinates, y_coordinates)), np.int32)
    cv2.fillPoly(mask, [polygon], (255, 255, 255))
    masked_image = cv2.bitwise_and(resized_image, mask)
    os.remove(os.path.join(directory, filename))
    idx = int(filename.split("_")[1].split(".")[0])
    cv2.imwrite(os.path.join(directory, f"Layer_{idx+1}.jpg"), masked_image)

    print(f"Saved Layer_{idx+1}.jpg")
```
